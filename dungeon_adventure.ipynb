```python
# Step 1: Player Setup
# Prompt the player for their name.
player_name = input("Enter you adventure name: ")


# Set health to 10.
health = 10
print(f"\nWelcome, {player_name}! You begin your adventure with {health} health.\n")

# Create an empty list to store their inventory.
inventory = []
```
